```python
# Importing necessary packages 

# For all problems
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
import numpy.random as rand

# For question 3
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Problem 1. (a)

b0_range = np.linspace(-2,2, 41)
b1_range = np.linspace(-2,2, 41)

b0_list = []
b1_list = []
s_val_list = []

for b0 in b0_range:
    for b1 in b1_range:
        s_val = 1/(1 + np.exp(-(b0+b1)))
        b0_list.append(b0)
        b1_list.append(b1) 
        s_val_list.append(s_val)
        
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(projection='3d')
ax.scatter(b0_list, b1_list, s_val_list)
ax.set_xlabel('b0') 
ax.set_ylabel('b1')
ax.set_zlabel('output')
ax.dist = 12
plt.show()
```

```python
# Problem 1. (b)

b0_range = np.linspace(-2,2, 41)
b1_range = np.linspace(-2,2, 41)

b0_list = []
b1_list = []
s_val_list = []

log_lik_list_sum = []

for b0 in b0_range:
    for b1 in b1_range:
        b0_list.append(b0)
        b1_list.append(b1)
        log_lik_list_sum.append(-np.log(1 + (np.exp(b0 + b1))) + -np.log(1 + (np.exp(-(b0 + b1)))))

        
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(projection='3d')
ax.scatter(b0_list, b1_list, log_lik_list_sum)
ax.set_xlabel('b0') 
ax.set_ylabel('b1')
ax.set_zlabel('log likelihood function')
ax.dist = 12
plt.show()
```

```python
# Problem 2

x = np.load('data.npy')
y = np.load('label.npy')

index_0 = np.where(y==0)[0][0]
index_1 = np.where(y==1)[0][0]

plt.imshow(x[index_0].reshape((28,28)))
plt.show()
plt.imshow(x[index_1].reshape((28,28)))
plt.show()
```

```python
# Problem 2 - Continued

rand.seed(19)

x_normalized = x/255
y_normalized = np.where(y==0, -1, y)

train_index = rand.choice(np.arange(0, 14780, 1), int(14780*.8), replace=False)
test_index = np.setdiff1d(np.arange(0, 14780, 1), train_index)

x_train = x_normalized[train_index]
y_train = y_normalized[train_index]
x_test = x_normalized[test_index]
y_test = y_normalized[test_index]
```

Random splitting is better than sequentially splitting here because the data is ordered, the earlier instances are all 0s, the later instances are all 1s.

```python
# Problem 2 - Continued 

def compute_loss(data, labels, b, b0):
    
    logloss = np.sum(np.log(np.exp((np.dot(data, b)+b0) * -labels) + 1))/len(labels)

    return logloss

def compute_gradients(data, labels, b, b0):
    
    b0_grad = -np.sum(
        ((np.exp(-labels*(b0 + np.dot(data, b))))/(
            np.exp(-labels*(b0 + np.dot(data, b))) + 1)) * labels)/len(labels)

    b_grad = -(np.dot((((np.exp(-labels*(b0 + np.dot(data, b))))/(
            np.exp(-labels*(b0 + np.dot(data, b))) + 1)) * labels), data))

    b_grad = b_grad/len(data)

    
    
    return b0_grad, b_grad
```

```python
# Problem 2 - Continued 

rand.seed(19)

b = rand.normal(size=784)
b0 = rand.normal()
lr = .05

iteration = []
test_acc = []
train_losses = []
test_losses = []

for i in range(50):
    loss = compute_loss(x_train, y_train, b, b0)
    
    b0_grad, b_grad = compute_gradients(x_train, y_train, b, b0)
    
    b = b - lr*b_grad
    b0 = b0 - lr*b0_grad
    
    test_loss = compute_loss(x_test, y_test, b, b0)
    
    test_prob = 1/(1+np.exp(-(b0 + np.dot(x_test, b))))
    test_pred = np.where(test_prob>=.5, 1, -1)
    test_accuracy = np.sum(np.where((y_test * test_pred)>0, 1, 0))/len(y_test)
    
    iteration.append(i+1)
    test_acc.append(test_accuracy)
    test_losses.append(test_loss)
    train_losses.append(loss)
    
    
#     print('Iteration ' + str(i+1) + ':')
#     print('Test Loss: ' + str(test_loss))
#     print('Test Accuracy: ' + str(test_accuracy))
#     print()
    
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot()
ax.plot(iteration, test_acc)
plt.xlabel('Iteration Number')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. Iterations')
plt.show()

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot()
ax.plot(iteration, train_losses, label='Train Loss')
ax.plot(iteration, test_losses, label='Test Loss')
plt.xlabel('Iteration Number')
plt.ylabel('Loss')
plt.title('Loss vs. Iterations')
plt.legend()
plt.show()
```

```python
# Problem 3 

rand.seed(19)

def generate_data(size):
    x = []
    y = []
    for i in range(0, size):
        num = rand.uniform(0,1)
        x.append([num])
        if num>=.5:
            if num>=.75:
                y.append(rand.choice([1,0], p=[.15, .85]))
            else:
                y.append(rand.choice([1,0], p=[.85, .15]))
        else:
            if num<=.25:
                y.append(rand.choice([1,0], p=[.15, .85]))
            else:
                y.append(rand.choice([1,0], p=[.85, .15]))
            
    return x, y

def bayes_optimal_classifier_pred(data):
    pred = []
    for x in data:
        if np.sum(x)>=.5:
            if np.sum(x)>=.75:
                pred.append(0)
            else:
                pred.append(1)
        else:
            if np.sum(x)<=.25:
                pred.append(0)
            else:
                pred.append(1)
    
    return pred
                
                
# With 100 data points
x_train, y_train = generate_data(100)
x_test, y_test = generate_data(100)

clf = LogisticRegression()
model = clf.fit(x_train,y_train)
y_pred = model.predict(x_test)
print('Logistic Regression Accuracy for 100 data points: ' + str(accuracy_score(y_test, y_pred)))
print('Bayes Optimal Classifier Accuracy for 100 data points: ' 
      + str(accuracy_score(bayes_optimal_classifier_pred(x_test), y_test)))
print()

# With 1000 data points
x_train, y_train = generate_data(1000)
x_test, y_test = generate_data(1000)

clf = LogisticRegression()
model = clf.fit(x_train,y_train)
y_pred = model.predict(x_test)
print('Logistic Regression Accuracy for 1000 data points: ' + str(accuracy_score(y_test, y_pred)))
print('Bayes Optimal Classifier Accuracy for 1000 data points: ' 
      + str(accuracy_score(bayes_optimal_classifier_pred(x_test), y_test)))
```

Question 3 Written Answers:

Since our data is not distributed logistically, our logistic regression model is not going to work well as a classifier for this data set. However, since we know the distribution, we can use Bayes' optimal classifier to predict a label for a given data point by picking the label with the highest probability. For this distribution, there is a 50/50 split between 0 and 1 labels, and an 85% chance of generating one of those labels for various different ranges of x. Therefore, for each range of x that our data point falls into, we can predict the label with the highest probability every time, which will lead us to be right about 85% of the time. Since we know the distribution, the Bayes' optimal classifier will always perform better than any other classifier, as shown by comparing the accuracy to that of logistic regression, which performs poorly here because the data is not distributed logistically. The results for both a sample size of 100 and a sample size of 1000 are similar because the logistic regression model will not be able to learn how to predict this distribution well, even with an increased sample size. Increasing the sample size will just lead to the classifiers converging to .5 for the logistic regression, since it is essentially randomly guessing, and .85 for the Bayes' classifier, since that is the probability the distribution generates a given label.
